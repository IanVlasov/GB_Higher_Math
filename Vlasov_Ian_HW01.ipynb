Задание 2.3

```python
def draw_circle():
    t = np.linspace(0, 2*np.pi, 100)
    R = float(input('Введите радиус: '))
    plt.plot(R*np.cos(t), R*np.sin(t))
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_circle()
```

```python
def draw_ellipse():
    t = np.linspace(0, 2*np.pi, 100)
    Rx = float(input('Введите радиус по оси x: '))
    Ry = float(input('Введите радиус по оси y: '))
    plt.plot(Rx*np.cos(t), Ry*np.sin(t))
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_ellipse()
```

```python
def draw_hyperbola():
    x = np.linspace(-10, 10, 100)
    a = float(input('Введите параметр a: '))
    b = float(input('Введите параметр b: '))
    y = np.sqrt(((x**2)/(a**2)-1)*b**2)
    plt.plot(x, y)
    plt.plot(x, -y)
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_hyperbola()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 3*Y + 50
Z2 = 2*X + 3*Y
ax1.plot_wireframe(X, Y, Z1)
ax1.plot_wireframe(X, Y, Z2)
```

```python
fig2 = plt.figure()
fig3 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax3 = fig3.add_subplot(111, projection='3d')
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z3 = np.sqrt(10*X**2 + 10*Y**2 - X*Y)
Z4 = (50*X**2 - 50*Y**2 - 10*X + X*Y)
ax2.plot_surface(X, Y, Z3)
ax3.plot_surface(X, Y, Z4)
```
