## Элементы теории вероятностей

### 1. Задание (теорема сложения)

Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

\begin{equation}
p_{2} = 1/6
\end{equation}
\begin{equation}
p_{5} = 1/6
\end{equation}
\begin{equation}
p_{2,5} = p_{2} + p_{5} = 1/3
\end{equation}

### 2. Задание (теорема умножения)

Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.


\begin{equation}
p_{2,5} = p_{2} * p_{5} = 1/36
\end{equation}

### 3. Задание 

Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.

Не уверен, что правильно понял условие. Думаю речь идет о том, что в первом случае нас устраивает выпадение 2 или 5 очков (вероятность 1/3, как показано выше). При втором подбрасывании необходимо, чтобы выпала 2 при условии, что в первом испытании выпала 5, либо чтобы выпала 5, при условии, что в первом испытании выпала 2. Выпадение определенного числа - независимое событие, в первом случае нас устраивает 2 исхода из 6, во втором только 1. Тогда вероятность итогового события А:
\begin{equation}
p_{A} = 2/6 * 1/6 = 1/18
\end{equation}

### 4. Задание  (Геометрическая вероятность +интервалы)

На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

Чтобы удовлетворить обоим условиям, точка должна попасть в отрезок [5, 9]. Длина которого равна 4. Тогда вероятность того, что точка С попадет в нужный промежуток:
\begin{equation}
p_{С} = 4/20 = 1/5
\end{equation}

### 5. Задание

Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882227?

В формировании номера могут участвовать 10 цифр на каждой из 7 позиции. Цифры могут повторяться. Тогда общее количество возможных номеров определяются формулой количества размещений с повторениями:
\begin{equation}
A_{10}^{7}=10^{7}
\end{equation}
При этом нас устраивает только один случай из всей выборки. Следовательно вероятность будет определяться по формуле:
\begin{equation}
p_{8882227}=1/10^{7}
\end{equation}


### 6. Задание

Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

Здесь необходимо успользовать формулу количества размещений без повторений (сколькими способами можно выбрать из 9 цифр две и в каждой выборке переставить их местами):
\begin{equation}
A_{9}^{2}=8*9=72
\end{equation}
И так как из 72 возможных номеров абонента устраивает только один, то вероятность такого события равна:
\begin{equation}
p_{true}=1/72
\end{equation}

### 7. Задание** (необязательное) 

Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?

Это куб со сторонами 3*3*3. если его покрасили снаружи белой краской, то:
* существует 6 кубиков с только 1 белой гранью
* 12 кубиков с 2 соседними белыми гранями
* 8 кубиков с 3 соседними белыми гранями
* 1 полностью черный кубик

Допустим, что все кубики пронумерованы.

Так же будем считать, что грани не пронуменорованы и имеют различия только в цвете, таким образом кубик со всеми черными гранями можно расположить единственным способом, а, например с 1 белой гранью только 6 различными способами, в зависимости от расположения единственной белой грани.

Так же считаем, что кубики выставляют по очереди в следующем порядке: нет белых граней - 1 белая грань - 2 белых грани - 3 белых грани. Возможных позиций с выставлением каждого нового кубика становится меньше.

Вероятность того, что полностью черный кубик встанет на свое место 1/27

Вероятность того, что все 6 кубиков с 1 белой гранью встанут на правильное место, при условии, что черный уже стоит верно:
\begin{equation}
p_{1} = (1/6)^{6}*(6/26*5/25..*1/21) = (1/6)^{6}*6!*20!/26!
\end{equation}

Вероятность того, что 12 из 20 кубиков с 2 свободными гранями встанут на свое место (расположить такой кубик можно 12 различными способами):
\begin{equation}
p_{2} = (1/12)^{12}*(12/20*11/19..*1/9) = (1/12)^{12}*12!*8!/20!
\end{equation}

Вероятность того, что 8 оставшихся кубиков встанут на свое место (8 различных расположений):
\begin{equation}
p_{3} = (1/8)^{8}
\end{equation}

Итого:

\begin{equation}
p_{all-white} = p_{1}*p_{2}*p_{3} = \frac{6!\cdot12!\cdot8!}{26!\cdot6^{6}\cdot8^{8}\cdot12^{12}}
\end{equation}
