```python
def vector_length(a):
    sum = 0
    for item in a:
        sum += item*item
    sum = np.sqrt(sum)
    return sum

vector_length(b)
```

```python
def draw_circle():
    t = np.linspace(0, 2*np.pi, 100)
    R = float(input('Введите радиус: '))
    plt.plot(R*np.cos(t), R*np.sin(t))
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_circle()
```

```python
def draw_ellipse():
    t = np.linspace(0, 2*np.pi, 100)
    Rx = float(input('Введите радиус по оси x: '))
    Ry = float(input('Введите радиус по оси y: '))
    plt.plot(Rx*np.cos(t), Ry*np.sin(t))
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_ellipse()
```

```python
def draw_hyperbola():
    t = np.linspace(-2, 2, 100)
    a = float(input('Введите параметр a: '))
    b = float(input('Введите параметр b: '))
    x = a*np.cosh(t)
    y = b*np.sinh(t)
    plt.plot(x, y)
    plt.plot(-x, y)
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    
draw_hyperbola()
```

```python
def convert_polar_to_cartesian(r, fi_grad):
    fi_rad = (fi_grad * math.pi) / 180
    y = r * np.sin(fi_rad)
    x = r * np.cos(fi_rad)
    cartesian = np.array([x, y])
    return cartesian
```

```python
from scipy.optimize import brentq

x = linspace(-2, 5., 1000)
y = np.exp(x) - x**3 + 2*x - 1
plt.plot(x, y)
plt.grid()
```

В уравнение 2 был подставлен у и 1 была перенесена в левую часть

Нули функции (на графике видно, что функция пересекает ось абсцисс в 4 местах):

```python
nulls = np.array([])

nulls = np.append(nulls, brentq(lambda x: np.exp(x) - x**3 + 2*x - 1, -2, -1))
nulls = np.append(nulls, brentq(lambda x: np.exp(x) - x**3 + 2*x - 1, -0.5, 0.5))
nulls = np.append(nulls, brentq(lambda x: np.exp(x) - x**3 + 2*x - 1, 2, 3))
nulls = np.append(nulls, brentq(lambda x: np.exp(x) - x**3 + 2*x - 1, 4, 5))

print(nulls)
```

```python
y_solution = nulls**2 - 1
print(y_solution)
```
